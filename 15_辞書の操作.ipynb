## 【課題1】山口県のみピックアップ
以下のように，「市名と県名の組み合わせ」を格納した辞書
cities がある。
この辞書から「山口県に属する市の名前のみ」を取り出すコードを書け。

```
cities = {"宇部市":"山口県", "北九州市":"福岡県", "福岡市":"福岡県", "周南市":"山口県",
          "防府市":"山口県", "大宰府市":"福岡県", "佐世保市":"長崎県", "西海市":"長崎県",
          "山陽小野田市":"山口県", "下関市":"山口県"}
```

````python
cities = {"宇部市":"山口県", "北九州市":"福岡県", "福岡市":"福岡県", "周南市":"山口県",
          "防府市":"山口県", "大宰府市":"福岡県", "佐世保市":"長崎県", "西海市":"長崎県",
          "山陽小野田市":"山口県", "下関市":"山口県"}
````

````python
# パターン１：keys() メソッドを用いて，辞書から key を取り出す形式
for city in cities.keys():
    if cities[city] == "山口県":
        print(city)
````

````python
# パターン２：items() メソッド用いて，辞書から key, value のペアを取り出す形式
for city, pref in cities.items():
    if pref == "山口県":
        print(city)
````

(解説)辞書内の値(value)を順次，チェックしながら，「山口県」に一致するキーのみを取り出す，という方針。

次のように書くと，辞書 `cities` 内のキーを順番に取り出せる。キーの中身は `city` へ入る。
```
for city in cities.keys():
```
こうやってキーのみを取り出した場合，値へアクセスするには `cities[city]` と書くことになる。

「なんかメンドクセーな」と感じるならば，後述の `items()` メソッドがお薦めである。`items()` メソッドは「キーと値のペア」を順次，取り出するため，次のように「それぞれの中身を受け取る変数」を２個，準備する必要がある。
```
for city, pref in cities.items():
```


(余談) 最近の Python 言語では，次のように keys() メソッドを用いずにキーの一覧を取得する書き方が推奨されている。  
しかし，個人的には明示的に `keys()` を付与する方が分かりやすいと思う。

````python
for k in cities:
    print(k)
````

### 課題 2-1 : 以下のように，各点数を取った人数を知りたい。
```
9点:1人
20点:2人
.........
70点:3人
71点:9人

.........
98点:2人

```

````python
scores=[53, 79, 63, 47, 86, 79, 75, 72, 97, 73, 68, 71, 49, 64, 65, 41, 75, 62, 63, 66, 97, 90, 71, 45, 65, 78,
	61, 90, 71, 72, 66, 51, 89, 52, 66, 62, 90, 80, 74, 48, 83, 49, 69, 78, 46, 49, 67, 95, 91, 94, 50, 94,
	51, 54, 71, 88, 78, 78, 38, 68, 68, 46, 63, 62, 55, 20, 71, 40, 75, 28, 38, 41, 78, 60, 39, 55, 55, 43,
	80, 38, 45, 70, 34, 34, 50, 79, 71, 72, 26, 75, 74, 93, 55, 72, 60, 95, 68, 68, 55, 91, 59, 84, 54, 67,
	39, 79, 23, 76, 59, 66, 53, 71, 47, 72, 23, 86, 92, 77, 70, 73, 31, 49, 77, 42, 96, 55, 95, 76, 30, 83,
	29, 73, 9, 85, 73, 69, 57, 51, 59, 79, 97, 67, 65, 39, 85, 74, 70, 71, 100, 73, 56, 60, 51, 98, 81, 71,
	47, 55, 51, 65, 64, 66, 39, 62, 45, 20, 77, 45, 84, 59, 44, 55, 53, 59, 83, 84, 98, 76, 40, 75, 50, 60,
	41, 55, 54, 74, 38, 79, 44, 57, 77, 64, 58, 81, 87, 69, 52, 63, 42, 73]
````

````python
counter = dict() # 各点数の人数を保存する辞書

# リスト scores から値を取り出し，それぞれの値の出現回数を記録していく。
for score in scores:
    if score in counter: # 既に点数 score が現れていれば
        counter[score] += 1
    else: # 点数 score が初めて現れる場合
        counter[score] = 1

# 辞書のソート
# (注意) sorted_counter は「タプルのリスト」となる。
sorted_counter = sorted(counter.items())

# sorted_score から「点数，人数」の組を順次，取り出して表示
for score, count in sorted_counter:
    print("{}点：{}人".format(score, count))
````

(解説) 基本的に，テキストに掲載された「アンケート集計プログラム」をまるっとぱくればよい。

また，上記のコードには「辞書のソート」が含まれている。
辞書のソートには「キーによるソート」と「値によるソート」があり，上記の要求においては「キーによるソート」が必要となる。

「リストのソート」では，sort() メソッドまたは sorted() 関数を用いた。辞書の場合も sorted() 関数を使えばよいが，その引数は (`conuter` でなく）`counter.items()` を渡す点に注意が必要だろう。
```
# 辞書のソート
# (注意) sorted_counter は「タプルのリスト」となる。
sorted_counter = sorted(counter.items())
```


### 課題 2-2 : 以下のように 20点台が 7人，30点台が 12人・・・という風に，度数分布表を作りたい。
```
0点台: 1人
20点台: 7人
30点台: 12人
40点台: 24人
50点台: 34人
60点台: 36人
70点台: 50人
80点台: 17人
90点台: 18人
100点台: 1人
```

````python
counter = dict() # 各点数の人数を保存する辞書
for score in scores:
    # 点数を丸める
    score = (score // 10) * 10

    if score in counter: # 既に点数 score が現れていれば
        counter[score] += 1
    else: # 点数 score が初めて現れる場合
        counter[score] = 1

# 辞書のソート
# (注意) sorted_counter は「タプルのリスト」となる。
sorted_counter = sorted(counter.items())

# sorted_score から「点数，人数」の組を順次，取り出して表示
for score, count in sorted_counter:
    print("{}点：{}人".format(score, count))
````

(解説) 上記にコードにおけるポイントは「点数を丸める」
```
score = (score // 10) * 10
```
にある。
いくつかの例を挙げると
```
5点 → 0点台
10点 → 10点台
18点 → 10点台
92点 → 90点台
96点 → 90点台
```
という形で置き換えればよい。具体的には以下の「２段階の操作」となる。
```
10(点) → 1 → 10
18(点) → 1 → 10
92(点) → 9 → 90
```

18 を 10 で割り，その商 1 を得るための演算子は `//` である。単純な割り算 `/` を用いると `1.8` が出てしまう。

## (参考) 出現回数を数えるには collections クラスが便利
上記では，「辞書の機能を理解する」という観点から，「特定の値の出現回数」を数えるため `for` ループを回した。
一方，`collections` クラスを用いると `Counter()` メソッド一発で処理できる。

以下の例では「出現回数を数える `Counter()` メソッド」のみ示しているが，これ以外にどのようなメソッド（＝機能）が実装されているか，調べてみるとよいだろう。

````python
import collections
c = collections.Counter(nums)
for num, count in c.items():
    if count == 2:
        print(num)
````
